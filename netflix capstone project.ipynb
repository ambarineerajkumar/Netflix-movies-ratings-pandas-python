# Q5: Create any plot to visualize the distribution of your choice . Which Seaborn function will
 # you use, and what information can you obtain from the plot

```python
plt.figure(figsize=(20,8))
target_col = "View Rating"  
sns.countplot(x=df[target_col])
plt.title("Distribution of View Ratings")
plt.xticks(rotation=45)
plt.show()
```

 # Q6: Visualize the distribution of the 'Runtime' column using a Seaborn plot with specific¬∂
 # parameters. What insights can you gather from this visualization?

```python
# Visualizing the distribution of 'Runtime'
sns.histplot(df['Runtime'], bins=30, kde=True)
plt.title("Distribution of Runtime")
plt.xlabel("Runtime (minutes)")
plt.show()
```

# Q7: Create any plot of choice to explore the relationship between two numerical columns. 
 # Which Seaborn function should you use, and what does the scatter plot reveal?

```python
# Scatter plot between IMDb Score and Runtime
sns.scatterplot(x=df['IMDb Score'], y=df['Runtime'])
plt.title("IMDb Score vs. Runtime")
plt.xlabel("IMDb Score")
plt.ylabel("Runtime (minutes)")
plt.show()
```

 #   Q8: Introduce a different type of visualization to gain more insights. What Seaborn or¬∂

```python
# Additional visualization (Box Plot for IMDb Scores)
sns.boxplot(y=df['IMDb Score'])
plt.title("Box Plot of IMDb Scores")
plt.show()
```
